## Tensor Reshape and Slicing

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and View
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
# Reshape if we don't know number of items
my_torch2 = torch.arange(10)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(2,-1)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch3.view(2,5)
my_torch4
```

```python
# Slices
my_torch5 = torch.arange(10)
my_torch5
```

```python
for i in range(4):
  my_torch5[4] = 12+i
my_torch5
```

## Tensor Math

```python
tensor1 = torch.tensor([1,2,3,4])
tensor2 = torch.tensor([5,6,7,8])
```

Addition

```python
tensor1 + tensor2
```

```python
torch.add(tensor1,tensor2)
```

Subtraction

```python
tensor2 - tensor1
```

```python
torch.subtract(tensor2,tensor1)
```

Multiply

```python
tensor1 * tensor2
```

```python
torch.multiply(tensor1,tensor2)
```

Divide

```python
torch.divide(tensor2,tensor1)
```

```python
torch.mul(tensor1,tensor2)
```

Modulo

```python
tensor2 % tensor1
```

```python
torch.remainder(tensor2,tensor1)
```

Power

```python
torch.pow(tensor1,tensor2)
```

```python
tensor1.add_(tensor2)
```
