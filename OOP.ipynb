## Blick/ Blick Switch Classes

```python
import time
class Blick:
  def __init__(self,name,ammo):
    self.name = name
    self.ammo = ammo

  def shoot(self):
    while self.ammo > 0:
       if input() == "g":
        self.ammo -= 1
        print(f'{self.name} ammo: {self.ammo}')

class Blick_Switch(Blick):
  def switch(self):
    while self.ammo > 0:
      time.sleep(0.2)
      self.ammo -= 1
      print(f'{self.name} ammo: {self.ammo}')


revolver = Blick("revolver", 6)
glock = Blick_Switch("glock", 17)

revolver.shoot()
```

## Spider-man Classes

```python
class Spider_Man:
  def __init__(self, name, web_cartridge):
    self.name = name
    self.web_cartridge = web_cartridge

  def web_shoot(self):
    while self.web_cartridge > 0:
      if input() == "s":
        self.web_cartridge -= 1
        print(f'{self.name} web cartridge: {self.web_cartridge}')

class Miles_Morales(Spider_Man):
  def venom_punch(self):
    if input() == "p":
      print(f'{self.name} Venom Punch!')


class Peter_Parker(Spider_Man):
  def spider_arms(self):
    if input() == "a":
      print(f'{self.name} Spider Arms Activated!')

  def web_shoot(self):
    while self.web_cartridge > 0:
      if input() == "s":
        self.web_cartridge -= 1
        print(f'{self.name} web cartridge: {self.web_cartridge}')

miles = Miles_Morales("Miles Morales", 10)
peter = Spider_Man("Peter Parker", 10)

miles.venom_punch()
```
